```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies = pd.read_csv(r"C:\Users\Utente\Desktop\Nuova cartella\imdb_movies.csv")
movies.head()
```

```python
pd.set_option('display.max.rows',11000)
pd.set_option('display.max.columns',22)
```

```python
movies.info()
```

```python
movies.head()
```

```python
#Rimuovere i duplicati

movies.drop_duplicates(inplace = True)
```

```python
#Rimuovere dati nulli colonna 'generi'

movies.dropna(subset=['genres'], inplace = True)
```

```python
movies['profit'] = movies['revenue'] - movies['budget']
```

```python
movies.head()
```

```python
#Creare un dataset più piccolo con solo i dati che che interessano
```

```python
movies_genre = movies[['popularity', 'budget','revenue','original_title','runtime','genres','release_date','vote_count','vote_average','profit']]
```

```python
movies_genre.head()
```

```python
from pandas import Series, DataFrame

```

```python
#I dati sul genere risultano inutilizzabili allo stato attuale. 
#Divido il contenuto in altrettanti valori indice assegnati allo stesso film, così da  renderli raggruppabili singolarmente
split = movies_genre['genres'].str.split('|').apply(Series,1).stack()

split.index = split.index.droplevel(-1)
split.name = 'genre_split'
del movies_genre['genres']
movies_genre = movies_genre.join(split)
```

```python
movies_genre
```

### Research Questions:

#### 1. Quale genere risulta essere il più comune(numero di film prodotti) ?

### Diamo un'occhiata all'andamento dei profitti negli anni

```python
#Ricreo il dataset però con la 'release year'
```

```python
movies_genre = movies[['popularity', 'budget','revenue','original_title','runtime','genres','release_year','vote_count','vote_average','profit']]
```

```python
split = movies_genre['genres'].str.split('|').apply(Series,1).stack()

split.index = split.index.droplevel(-1)
split.name = 'genre_split'
del movies_genre['genres']
movies_genre = movies_genre.join(split)
```

```python
movies_genre.head()
```

```python
time_genre = pd.DataFrame(movies_genre.groupby(['release_year','genre_split'])['profit'].mean())

time_genre
```

```python
final_genre = pd.pivot_table(time_genre,values ='profit', index = ['genre_split'], columns = ['release_year'])
```

```python
sns.set(rc = {'figure.figsize': (15,10)})
sns.heatmap(final_genre, cmap = 'YlGnBu', linewidths = .5 )
plt.title('Genres by Profit per Year')
```

- I film di genere 'fantasy' e 'family' hanno riscosso sempre più successo dagli inizi degli anni 2000. Senzo puzza di 'Marvel'. Anche quelli di 'animazione' sono andati migliorando.
- I film 'adventure' sono andati sempre molto bene.
